```python
#Import das bibliotecas

import pandas as pd
import numpy as np
import re
```

```python
#Ler o arquivo
#update df set Situação = trim(Situação)

df = pd.read_csv("acre_boladao_JAN.csv", sep=',')
df['id'] = range(len(df))
df['Situação'] = df['Situação'].str.strip()
df
```

```python
#Copiar para um novo dataframe os que estão a coluna SITUAÇÃO errada

df_tratar = df[df['Líquido'].isnull()]
df_tratar
```

```python
#Realocar as colunas paras as posições certas

df_tratar['Líquido'], df_tratar['Bruto'], df_tratar['Desconto'], df_tratar['Cargo'] = df_tratar['Desconto'], df_tratar['Cargo'], df_tratar['Bruto'], df_tratar['Situação']
df_tratar
```

```python
#Tratar a Coluna Cargo

palavras_a_remover_0 = ['APOSENTADO', 'EM EXERCÍCIO', 'EM FÉRIAS', 'PENSIONISTA', 'AFASTADO/LICENCIADO']
for palavra in palavras_a_remover_0:
    df_tratar['Cargo'] = df_tratar['Cargo'].str.replace(f'^{palavra}', '', regex=True)

df_tratar
```

```python
#Tratar a coluna Situação

palavras_a_remover_1 = df_tratar['Cargo']

# Aplicar a remoção das palavras à coluna 'Situação'
for palavra in palavras_a_remover_1:
    df_tratar['Situação'] = df_tratar['Situação'].str.replace(fr'\b{palavra}\b', '', regex=True, case=False).str.strip()
df_tratar
```

```python
df_tratar_2 = df_tratar[df_tratar['Líquido'].isnull()]
df_tratar_2['Líquido'], df_tratar_2['Bruto'], df_tratar_2['Desconto'], df_tratar_2['Cargo'] = df_tratar_2['Desconto'], df_tratar_2['Cargo'], df_tratar_2['Bruto'], df_tratar_2['Situação']
df_tratar_2
```

```python
df_tratar_2['Situação'] = df_tratar_2['Nome']
df_tratar_2
```

```python
palavras_a_remover_3 = pd.read_csv('cargos_servidor_0.csv')
palavras_a_remover_3=palavras_a_remover_3['Cargo'].to_list()

#Tratar a coluna Nome

palavras_a_manter_2 = ['APOSENTADO', 'EM EXERCÍCIO', 'EM FÉRIAS', 'PENSIONISTA', 'AFASTADO/LICENCIADO', 'EXONERADO/RESCISO', 'CEDIDO']

def manter_apenas_palavras(nome):
    for palavra in palavras_a_remover_3:
        index = nome.find(palavra)
        if index != -1:
            return nome[:index + len(palavra)].strip()
    return nome


df_tratar_2['Nome'] = df_tratar_2['Nome'].apply(manter_apenas_palavras)
df_tratar_2
```

```python
#excluir a coluna ID e salvar o arquivo Finalizado

df.drop(columns=['id'], inplace=True)

df.update(df_tratar)

df.to_csv('tratamento_1.csv', index=False, sep=';')
```
