```python
#Import das bibliotecas

import pandas as pd
import numpy as np
import re
```

```python
#Ler o arquivo
#update df set Situação = trim(Situação)

df = pd.read_csv("Bases/acre_boladao_FEV_2023.csv", sep=',')
df['id'] = range(len(df))
df['Situação'] = df['Situação'].str.strip()
df
```

```python
#Copiar para um novo dataframe os que estão a coluna SITUAÇÃO errada

df_tratar = df[df['Líquido'].isnull()]
df_tratar
```

```python
#Realocar as colunas paras as posições certas

df_tratar['Líquido'], df_tratar['Bruto'], df_tratar['Desconto'], df_tratar['Cargo'] = df_tratar['Desconto'], df_tratar['Cargo'], df_tratar['Bruto'], df_tratar['Situação']
df_tratar
```

```python
#Tratar a Coluna Cargo

palavras_a_remover_0 = ['APOSENTADO', 'EM EXERCÍCIO', 'EM FÉRIAS', 'PENSIONISTA', 'AFASTADO/LICENCIADO']
for palavra in palavras_a_remover_0:
    df_tratar['Cargo'] = df_tratar['Cargo'].str.replace(f'^{palavra}', '', regex=True)

df_tratar
```

```python
#Tratar a coluna Situação

palavras_a_remover_1 = df_tratar['Cargo']

# Aplicar a remoção das palavras à coluna 'Situação'
for palavra in palavras_a_remover_1:
    df_tratar['Situação'] = df_tratar['Situação'].str.replace(fr'\b{palavra}\b', '', regex=True, case=False).str.strip()
df_tratar
```

```python
#excluir a coluna ID e salvar o arquivo Finalizado

df.drop(columns=['id'], inplace=True)

df.update(df_tratar)

df.to_csv('Bases/acre_boladao_FEV_2023_tratado.csv', index=False, sep=';')
```
