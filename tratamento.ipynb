```python
# Criando uma função para dividir a coluna
def corrigir_linhas_incorretas(row):
    # Verifica se há uma letra maiúscula na célula
    if any(c.isupper() for c in str(row['Situação'])):
        # Divide a célula usando a posição da primeira letra maiúscula
        return pd.Series([row['Situação'][:next(i for i, c in enumerate(row['Situação']) if c.isupper())],
                          row['Situação'][next(i for i, c in enumerate(row['Situação']) if c.isupper()):]])
    else:
        return row

# Aplica a função à coluna problemática
df[['Situação', 'Cargo']] = df.apply(corrigir_linhas_incorretas, axis=1)

df
```
